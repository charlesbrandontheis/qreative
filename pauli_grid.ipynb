```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Circle
import random
```

```python
def pauli_grid( rho, labels=False):
    
    box = {'ZI':(-1, 2),'XI':(-2, 3),'IZ':( 1, 2),'IX':( 2, 3),'ZZ':( 0, 3),'ZX':( 1, 4),'XZ':(-1, 4),'XX':( 0, 5)}
    
    for pauli in box:
        rho[pauli] = 1-2*random.random()
    
    # draw lines
    fig, ax = plt.subplots(figsize=[12,10])
    plt.plot( [1,3], [1,3], 'k-' )
    plt.plot( [3,0], [3,6], 'k-' )
    plt.plot( [0,-3], [6,3], 'k-' )
    plt.plot( [-3,-1], [3,1], 'k-' )
    plt.plot( [-1,2], [1,4], 'k-' )
    plt.plot( [-2,1], [4,1], 'k-' )
    plt.plot( [-1,2], [1,4], 'k-' )
    plt.plot( [-1,2], [5,2], 'k-', lw=0.5 )
    plt.plot( [-2,1], [2,5], 'k-', lw=0.5 )
   

    # draw circles
    for pauli in box:
        ax.add_patch( Circle(box[pauli], 0.65, color=(0.5,0.5,0.5), fill=False, lw=0.5 ) )
    
    # add bars
    for pauli in box:
        l = 1.15
        prob = (1+rho[pauli])/2
        if pauli in ['ZI','IZ','ZZ']:
            a = ( box[pauli][0], box[pauli][1]-l/2 )
            b = ( box[pauli][0], box[pauli][1]+(prob-0.5)*l )
            c = ( box[pauli][0], box[pauli][1]+l/2 )
            plt.plot( [a[0],b[0]], [a[1],b[1]], color=(0.0,0.0,1.0), lw=15 )
            plt.plot( [b[0],c[0]], [b[1],c[1]], color=(0.75,0.75,1.0), lw=15 )
        if pauli in ['XI','IX','XX']: 
            a = ( box[pauli][0]-l/2, box[pauli][1] )
            b = ( box[pauli][0]+(prob-0.5)*l, box[pauli][1] )
            c = ( box[pauli][0]+l/2, box[pauli][1] )
            plt.plot( [a[0],b[0]], [a[1],b[1]], color=(1.0,0.0,0.0), lw=15 )
            plt.plot( [b[0],c[0]], [b[1],c[1]], color=(1.0,0.75,0.75), lw=15 )
        if pauli in ['XZ','ZX']:
            a = ( box[pauli][0]-l/(2*np.sqrt(2)), box[pauli][1]-l/(2*np.sqrt(2)) )
            b = ( box[pauli][0]+(prob-0.5)*l/(np.sqrt(2)), box[pauli][1]+(prob-0.5)*l/(np.sqrt(2)) )
            c = ( box[pauli][0]+l/(2*np.sqrt(2)), box[pauli][1]+l/(2*np.sqrt(2)) )
            plt.plot( [a[0],b[0]], [a[1],b[1]], color=(0.0,1.0,0.0), lw=15 )
            plt.plot( [b[0],c[0]], [b[1],c[1]], color=(0.75,1.0,0.75), lw=15 )
    
    if labels:
        for pauli in box:
            plt.text(box[pauli][0]-0.05,box[pauli][1]-0.85, pauli)
    
    plt.axis('off')
    plt.show()
```

```python
pauli_grid({})
```
